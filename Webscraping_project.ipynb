```python
def drop_col(df):
  #we rename columns by removing ":" at the end
  col=list(df.columns)
  for i in range(len(df.columns)):
    if(df.columns[i].endswith(":")):
      col[i]=df.columns[i][:-1]
  df.columns=col
  temp=df.copy()
  for i in col:
    #we drop columns with more than 25% of null values, we already have a lot of columns
    if(df[i].isna().sum()>0.25*len(df)):
      temp=temp.drop([i],1)
  
  return temp

df=drop_col(res)
```

```python
#-- clean bluetooth column
#-- we replace 'oui' by mode of the bluetooth column and handle some cases
mod=dff["bluetooth"].mode()
dff["bluetooth"]=dff["bluetooth"].replace(["Oui","Oui, 4.0",np.nan],[mod.iloc[0],"4","Non"])
#dff["bluetooth"]=dff["bluetooth"].astype('float')
dff["bluetooth"].unique()
```

```python
#-- clean clavier column
dff["clavier"]=dff["clavier"].replace("Non","NC")
dff["clavier"]=dff["clavier"].fillna("NC")
dff["clavier"]=dff["clavier"].str.lower()
dff['clavier']=dff['clavier'].replace(["steelseries non-rétroéclairé","classique","standard","simplifié","alcantara"],"standard")
dff["clavier"][dff["clavier"].str.contains("rétroéclairé") &
               dff["clavier"].str.contains("mécanique")]="mécanique rétroéclairé"
dff["clavier"][dff["clavier"].str.contains("rétroéclairé") &
               (dff["clavier"].str.contains("mécanique")==False) & (dff["clavier"].str.contains("détachable")==False)]="rétroéclairé"
dff["clavier"][dff["clavier"].str.contains("rétroéclairage") ]="rétroéclairé"
dff["clavier"][dff["clavier"].str.contains("détachable") ]="détachable"
dff["clavier"].unique()
```

```python
#-- clean price column
#dff["price"]=dff["price"].str.split(",").str[0]
#dff["price"]=dff["price"].astype("int")
```

```python
#-- clean carte_graphique column
dff["temp"]=dff["carte_graphique"].str.upper().str.split().str[0:2]
dff["temp"]=dff["temp"].apply(lambda x: "_".join(x) if str(x)!='nan' else None)
dff["model_CG"]=dff["temp"].replace(["INTÉGRÉE_AU","PAS_DE"],["ORIGINE","NULL"])
dff["model_CG"]=dff["model_CG"].fillna("NULL")

dff["temp2"]=dff["carte_graphique"].str.upper().str.split().str[0:1]
dff["temp2"]=dff["temp2"].str[0]
dff["marque_CG"]=dff["temp2"].replace(["INTÉGRÉE","PAS"],["ORIGINE","NULL"])
dff["marque_CG"]=dff["marque_CG"].fillna("NULL")

dff=dff.drop(["temp","temp2","carte_graphique"],1,errors='ignore')
```

```python
#-- clean taille_ecran column
dff["tactile"]=dff["taille_ecran"].str.contains("Tactile")
dff["tactile"]=dff["tactile"].fillna(False)

dff["taille_ecran"]=dff["taille_ecran"].str.split().str[0] 
dff["taille_ecran"]=dff["taille_ecran"].str.replace('"','')

mod=dff["taille_ecran"].mode()
dff["taille_ecran"]=dff["taille_ecran"].replace(["Tactile","Double"],[mod.iloc[0],mod.iloc[0]])
dff["taille_ecran"]=dff["taille_ecran"].fillna(mod[0])
dff["taille_ecran"]=dff["taille_ecran"].str.replace(",",".")

dff["taille_ecran"]=dff["taille_ecran"].astype('float')
```

```python
#-- clean frequence column
dff["frequence"][dff["frequence"]>1000]=dff["frequence"]/1000 #some values are not in gHz
med=dff["frequence"].median()
dff["frequence"]=dff["frequence"].fillna(med)
```

```python
#-- clean dimensions column
dff["temp"]=dff["dimensions"].str.replace(" cm","")
dff["temp"]=dff["temp"].str.split(" x ")
dff["dimensions_1"]=dff["temp"].str[0].astype("float")
dff["dimensions_1"]=dff["dimensions_1"].fillna(dff["dimensions_1"].median())
dff["dimensions_2"]=np.nan
dff["dimensions_2"][dff["temp"].str[1].str.contains("NC")==False]=dff["temp"].str[1]
dff["dimensions_2"]=dff["dimensions_2"].astype("float")
dff["dimensions_2"]=dff["dimensions_2"].fillna(dff["dimensions_2"].median())
dff["dimensions_3"]=np.nan
dff["dimensions_3"][dff["temp"].str[2].str.contains("NC")==False]=dff["temp"].str[2]
dff["dimensions_3"]=dff["dimensions_3"].astype("float")
dff["dimensions_3"]=dff["dimensions_3"].fillna(dff["dimensions_3"].median())
dff=dff.drop(["temp","dimensions"],1)
```

```python
#-- clean lieu_fabrication column
mod=dff["lieu_fabrication"].mode()
dff["lieu_fabrication"]=dff["lieu_fabrication"].fillna(mod.iloc[0])
```

```python
#-- clean "lecteur_empreinte" column
dff["lecteur_empreinte"]=dff["lecteur_empreinte"].fillna("Non")
```

```python
#-- clean "lecteur_carte_memoire" column
dff["lecteur_carte_memoire"]=dff["lecteur_carte_memoire"].replace(["Non communiqué","Micro SD","SD","SD multiformat","4 en 1","3 en 1","2 en 1","SDHC"],
                                                                  ["Non",'Oui','Oui','Oui','Oui','Oui','Oui','Oui'])
dff["lecteur_carte_memoire"]=dff["lecteur_carte_memoire"].fillna("Non")
```

```python
#-- clean column RAM
#we fill null values with mode of RAM of each brand
dff["RAM"]=dff["RAM"].str.split().str[0]
k=dff.groupby(["brand"],as_index=False)["RAM"].agg(pd.Series.mode)
k["RAM"]=k["RAM"].str[0]
for i in dff["brand"].unique():
  val=k["RAM"][k["brand"]==i].iloc[0]
  dff["RAM"][dff["brand"]==i]=dff["RAM"][dff["brand"]==i].fillna(val)
dff["RAM"]=dff["RAM"].astype('int')
```

```python
#-- clean "nb_coeurs" column
#we fill null values with median of nb coeurs of each brand
k=dff.groupby(["brand"],as_index=False).agg({"nb_coeurs":"median"})
for i in dff["brand"].unique():
  val=k["nb_coeurs"][k["brand"]==i].iloc[0]
  dff["nb_coeurs"][dff["brand"]==i]=dff["nb_coeurs"][dff["brand"]==i].fillna(val)
```

```python
#-- clean norme_wifi column
dff["norme_wifi"]=dff["norme_wifi"].str.upper()
dff["norme_wifi"][dff["norme_wifi"].str.contains('AX',na=False)]="WIFI 6 (AX)"
dff["norme_wifi"][dff["norme_wifi"].str.contains('AC',na=False)]='WIFI 5 (N/AC)'
dff["norme_wifi"][dff["norme_wifi"].str.contains('WI-FI 6',na=False)]="WIFI 6 (AX)"
dff["norme_wifi"][(dff["norme_wifi"].str.contains('WIFI',na=False)==False) & (dff["norme_wifi"].str.contains('N',na=False))]="WIFI 4 (N)"
dff["norme_wifi"]=dff["norme_wifi"].fillna("NC")
dff["norme_wifi"].unique()
```

```python
#poids en kilo
dff["poids"]=dff["poids"].str.split().str[0]
dff["poids"]=dff["poids"].str.replace(',','.')
dff["poids"]=dff["poids"].astype('float')
dff["poids"][dff["poids"]>100]=dff["poids"]/1000 #some weights are in grams
med=dff["poids"].median()
dff["poids"]=dff["poids"].fillna(med)
```

```python
#-- clean processeur column
dff["processeur"]=dff["processeur"].str.upper()
dff["processeur"][dff["processeur"].str.contains('CORE I7',na=False)]='INTEL CORE I7'
dff["processeur"][dff["processeur"].str.contains('CORE I5',na=False)]='INTEL CORE I5'
dff["processeur"][dff["processeur"].str.contains('CORE I3',na=False)]='INTEL CORE I3'
dff["processeur"][dff["processeur"].str.contains('CORE I9',na=False)]='INTEL CORE I9'
dff["processeur"][dff["processeur"].str.contains('CELERON',na=False)]='INTEL CELERON'
dff["processeur"][dff["processeur"].str.contains('PENTIUM',na=False)]='INTEL PENTIUM'
dff["processeur"][dff["processeur"].str.contains('AMD RYZEN 7',na=False)]='AMD RYZEN 7'
dff["processeur"][dff["processeur"].str.contains('AMD RYZEN 3',na=False)]='AMD RYZEN 3'
dff["processeur"][dff["processeur"].str.contains('AMD RYZEN 5',na=False)]='AMD RYZEN 5'
dff["processeur"][dff["processeur"].str.contains('AMD RYZEN 9',na=False)]='AMD RYZEN 9'
dff["processeur"][dff["processeur"].str.contains('AMD',na=False) &(dff["processeur"].str.contains('RYZEN',na=False)==False)]='AMD'
dff["processeur"][dff["processeur"].str.contains('QUALCOMM',na=False)]='QUALCOMM'
dff["processeur"][dff["processeur"].str.contains('MEDIATEK',na=False)]='MEDIATEK'

#we fill null values with mode of processeur of each brand
k=dff.groupby(["brand"],as_index=False)["processeur"].agg(pd.Series.mode)
k["processeur"][k["processeur"].str.len()==2]=k["processeur"].str[0]
for i in dff["brand"].unique():
  val=k["processeur"][k["brand"]==i].iloc[0]
  dff["processeur"][dff["brand"]==i]=dff["processeur"][dff["brand"]==i].fillna(val)

dff["processeur"].str.upper().unique()
```

```python
#-- clean resolution column
dff["resolution1"]=dff["resolution"].str.split(" x ").str[0]
mod=dff["resolution1"].mode()
dff["resolution1"]=dff["resolution1"].fillna(mod[0]) #only one NaN
dff["resolution1"]=dff["resolution1"].astype("int")
dff["resolution2"]=dff["resolution"].str.split(" x ").str[1].str.split(" ").str[0]
mod=dff["resolution2"].mode()
dff["resolution2"]=dff["resolution2"].fillna(mod[0]) #only one NaN
dff["resolution2"]=dff["resolution2"].astype("int")
dff=dff.drop(["resolution"],1)
```

```python
#-- clean stockage column
def extract_ssd(row):
    if(row==np.nan):
        return np.nan
    else:
        words=row.split()
        for i in range(len(words)):
            if((words[i].upper()=="SSD") and ((i+1)<len(words))):
                return words[i+1]

def extract_hdd(row):
    if(row==np.nan):
        return np.nan
    else:
        words=row.split()
        for i in range(len(words)):
            if((words[i].upper()=="DUR") and ((i+1)<len(words))):
                return words[i+1]
def extract_emmc(row):
    if(row==np.nan):
        return np.nan
    else:
        words=row.split()
        for i in range(len(words)):
            if((words[i].upper()=="EMMC") and ((i+1)<len(words))):
                return words[i+1][:-2] #we remove 'Go' letters
            
dff["SSD"]=np.nan
dff["SSD"]=dff["stockage"][dff["stockage"].isnull()==False].apply(extract_ssd)
dff["HDD"]=dff["stockage"][dff["stockage"].isnull()==False].apply(extract_hdd)
dff["eMMC"]=dff["stockage"][dff["stockage"].isnull()==False].apply(extract_emmc)

dff[["SSD","HDD","eMMC"]] = dff[["SSD","HDD","eMMC"]].fillna("0")

dff["HDD"]=dff["HDD"].astype('int')
dff["SSD"]=dff["SSD"].astype('int')
dff["eMMC"]=dff["eMMC"].astype('int')

#-- we convert To in Go
dff["HDD"][dff["HDD"]<=5]=dff["HDD"]*1000
dff["SSD"][dff["SSD"]<=5]=dff["SSD"]*1000
dff["eMMC"][dff["eMMC"]<=5]=dff["eMMC"]*1000
dff[["HDD","SSD","stockage"]][((dff["HDD"]==0) & (dff["SSD"]==0) & (dff["eMMC"]==0))]
dff=dff.drop(["stockage"],1)
```

```python
#-- clean webcam column
dff["webcam"]=dff["webcam"].fillna("Non")
dff["webcam"].unique()
```

```python
#-- clean OS column
dff["OS"][dff["OS"].str.contains('Windows',na=False)]='Windows'
dff["OS"][dff["OS"].str.contains('Chrome',na=False)]='Chrome'
dff["OS"][dff["OS"].str.upper().str.contains('MAC',na=False)]='Mac'
dff["OS"][dff["OS"].str.contains('Pas de système',na=False)]='No OS'
dff['OS']=dff["OS"].fillna("No OS")
dff["OS"].unique()
```

```python
#-- clean type_resolution column
dff["type_resolution"][dff["type_resolution"].str.contains('UHD',na=False)]='UHD'
dff["type_resolution"][dff["type_resolution"].str.contains('2,2K',na=False)]='2K'
dff["type_resolution"][dff["type_resolution"].str.contains('QHD',na=False)]='QHD'
dff["type_resolution"][dff["type_resolution"].str.contains('Retina',na=False)]='Retina'
dff["type_resolution"][dff["type_resolution"].str.contains('WQ',na=False)]='WQXGA'
dff["type_resolution"][dff["type_resolution"].str.contains('Full HD+',na=False)]='Full HD'
mod=dff["type_resolution"].mode()
dff["type_resolution"]=dff["type_resolution"].fillna(mod[0])
dff["type_resolution"].unique()
```

```python
dff=dff.rename(columns={"dimensions_1":"longueur",
                        "dimensions_2":"hauteur",
                        "dimensions_3":"profondeur"})
```
