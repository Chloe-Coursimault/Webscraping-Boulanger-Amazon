# Retrieve caracteristics, description & reviews for each computer

```python
details_temp=[]
```

```python
for index, row in products.iterrows():
    if index>=len(details_temp):
        product=dict()
        product['link']=links[index]

        soup=BeautifulSoup(str(row.soup), "html.parser")
        description = soup.select('div.centerColAlign div.a-section ul.a-unordered-list > li > span.a-list-item')
        desc=""
        for result in description:
            desc+=result.get_text()
        desc.replace('\n','')
        product['description']=desc

        reviews = soup.select('div.reviews-content')
        review_grade = None
        review_title = None
        review_text = None
        grades = []
        titles=[]
        texts=[]
        for review in reviews:
            review_grade = review.select("a.a-link-normal")
            review_title = review.select("a.review-title")
            review_text = review.select("div.reviewText")
            for a in review_grade:
                if 'étoiles' in str(a):
                    grades.append(str(a.get('title')).split(' ')[0])
            for a in review_title:
                titles.append(a.get_text())
            for a in review_text:
                texts.append(a.get_text())
        product['grades']=grades
        product['titles']=titles
        product['reviews']=texts

        dtls = soup.select('div.a-expander-content table.a-normal.a-spacing-micro tr')
        for detail in dtls:
            product[detail.select("span.a-size-base")[0].get_text()]=detail.select("span.a-size-base")[1].get_text()

        details_temp.append(product)
        #break
```

```python
def ram_trans(x):
    tempx=str(x).split(" ")
    if tempx[-1]=='MB':
        return str(0.032)
    if tempx[-1]=='Go':
        return str(tempx[0])
    else:
        return str(tempx[0])
```

```python
details['Taille de la mémoire RAM installée']=details['Taille de la mémoire RAM installée'].swifter.apply(ram_trans)
details['Taille de la mémoire RAM installée']=details['Taille de la mémoire RAM installée'].swifter.apply(float)
```

```python
def os_trans(x):
    tempx=str(x).split(" ")
    if tempx[0]=='Windows' or tempx[0]=="W11H" or tempx[0]=="Win" or tempx[0]=="W11PRO":
        return 'Windows'
    if tempx[0]=='Mac' or tempx[0]=='MacOS' or tempx[0]=='macOS':
        return 'Mac'
    if tempx[0]=='Chrome':
        return 'Chrome'
    else:
        return 'No OS'
    
    
```

```python
details["Système d'exploitation"]=details["Système d'exploitation"].swifter.apply(os_trans)
```

```python
details["Taille de l'écran"].unique()
```

```python
def taille_trans(x):
    tempx=str(x).split(" ")
    if tempx[-1]=='Pouces' or tempx[-1]=='pouces' or tempx[-1]=='Pieds':
        return str(tempx[0])
    elif tempx[-1]=="Centimètres":
        return str(17)
    else:
        return str(tempx[0])
```

```python
details["Taille de l'écran"]=details["Taille de l'écran"].swifter.apply(taille_trans)
```

```python
details["Taille de l'écran"]=details["Taille de l'écran"].swifter.apply(float)
```
